The choropleth map of Ibaraki pre.

```python
#モジュールをインポート
import geopandas as gpd
import pandas as pd
```

```python
# データを変数に入れる
ibaraki_ku = gpd.read_file('N03-23_08_230101.geojson')
```

```python
ibaraki_ku.plot()
```

```python
ibaraki_pop = pd.read_csv('ibaraki.csv')
```

```python
ibaraki_ku.rename(columns={'N03_007':'KEY_CODE'},inplace=True)
```

```python
ibaraki_ku['KEY_CODE'] = ibaraki_ku['KEY_CODE'].astype(int)
```

```python
ibaraki_ku_pop = ibaraki_ku.merge(ibaraki_pop,on='KEY_CODE')
```

```python
ibaraki_ku_pop
```

```python
ibaraki_ku_pop.info()
```

```python
ibaraki_ku_pop['人口総数'] = ibaraki_ku_pop['人口総数'].astype(int)
```

```python
ibaraki_ku_pop.plot(column='人口総数',legend=True)
```
